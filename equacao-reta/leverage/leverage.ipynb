```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import seaborn as sns
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 10
```



$$ h_{ii} = \frac{1}{n} + \frac{\left(x_{i} - \overline{x} \right)^{2}}{\sum_{i=1}^{n} \left(x_{i} - \overline{x} \right)^{2}}$$

onde:
- $n$ é o número total de amostras;
- $x_{i}$ é a i-ésima observação da amostra;
- $\overline{x}$ é a média da amostra;

$$critico = \frac{2\times n_{p}}{n}$$

onde:
- $n$ é o número total de amostras;
- $n_{p}$ é o número de parâmetros estimados pelo modelo, incluíndo o intercepto;

```python
concentracao = np.array([1, 2, 4, 6, 8, 10, 15, 20, 30, 100])
x_dff = np.square(concentracao - concentracao.mean())


Sxx = np.sum(x_dff)

leverage = 1/concentracao.size + x_dff/Sxx

limite_mild = 2*2/concentracao.size
limite_extreme = 3*2/concentracao.size
```

```python
plt.figure(figsize=(6,3))
x_aux = range(0,concentracao.size)
plt.scatter(concentracao, leverage, color='k', facecolor='none')


plt.axhline(y=limite_extreme, color="r", linestyle="--", label="Limite extremo")
plt.axhline(y=limite_mild, color="y", linestyle="--", label="Limite moderado")
plt.ylim(bottom=0)

plt.ylabel("Alavancagem")
plt.xlabel("Concentração")
plt.legend(bbox_to_anchor=(1.01,1))
plt.show()
```
