# Comparando os datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import seaborn as sns
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 10
```

```python
alfa = 0.05
```

```python
df_dataset_1 = pd.DataFrame({
    "Concentração": [1, 3, 5, 7, 9],
    "Sinal": [0.098, 0.135, 0.229, 0.285, 0.356],
    "Ordem de coleta": [1, 2, 3, 4, 5],    
    "Ensaio": ["Ensaio 1", "Ensaio 2", "Ensaio 3", "Ensaio 4", "Ensaio 5"],    
})
dataset_name_1 = "Dataset 1"
reg_linear_dataset_1 = stats.linregress(df_dataset_1['Concentração'], df_dataset_1['Sinal'])
b1_dataset_1 = reg_linear_dataset_1[0]
b0_dataset_1 = reg_linear_dataset_1[1]
df_dataset_1["ypred"] = b1_dataset_1*df_dataset_1['Concentração'] + b0_dataset_1
```

```python
df_dataset_2 = pd.DataFrame({
    "Concentração": [1, 3, 5, 7, 9, 5, 5],
    "Sinal": [0.098, 0.135, 0.229, 0.285, 0.356, 0.209, 0.249],
    "Ordem de coleta": [1, 2, 3, 4, 5, 6, 7],    
    "Ensaio": ["Ensaio 1", "Ensaio 2", "Ensaio 3", "Ensaio 4", "Ensaio 5", "Ensaio 6", "Ensaio 7",],    
})
dataset_name_2 = "Dataset 2"
reg_linear_dataset_2 = stats.linregress(df_dataset_2['Concentração'], df_dataset_2['Sinal'])
b1_dataset_2 = reg_linear_dataset_2[0]
b0_dataset_2 = reg_linear_dataset_2[1]
df_dataset_2["ypred"] = b1_dataset_2*df_dataset_2['Concentração'] + b0_dataset_2
```

```python
df_dataset_3 = pd.DataFrame({
    "Concentração": [1, 1, 1, 3, 3, 3, 5, 5, 5, 7, 7, 7, 9, 9, 9],
    "Sinal": [0.098, 0.087, 0.109, 0.125, 0.135, 0.145, 0.249, 0.229, 0.209, 0.290, 0.285, 0.280, 0.350, 0.356, 0.362],
    "Ordem de coleta": [1, 8, 11, 15, 2, 10, 7, 3, 6, 14, 4, 12, 13, 5, 9],    
    "Ensaio": ["Ensaio 1", "Ensaio 2", "Ensaio 3", "Ensaio 4", "Ensaio 5", "Ensaio 6", "Ensaio 7", "Ensaio 8", "Ensaio 9", "Ensaio 10", 
              "Ensaio 11", "Ensaio 12", "Ensaio 13", "Ensaio 14", "Ensaio 15"],    
})
dataset_name_3 = "Dataset 3"
reg_linear_dataset_3 = stats.linregress(df_dataset_3['Concentração'], df_dataset_3['Sinal'])
b1_dataset_3 = reg_linear_dataset_3[0]
b0_dataset_3 = reg_linear_dataset_3[1]
df_dataset_3["ypred"] = b1_dataset_3*df_dataset_3['Concentração'] + b0_dataset_3
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df_dataset_3['Concentração'], df_dataset_3['Sinal'], label=dataset_name_3, edgecolors='k', facecolor='None')
plt.plot(df_dataset_3['Concentração'], df_dataset_3['ypred'], label="Regressão", c='k', ls='--')
plt.xlabel('Concentração')
plt.ylabel('Sinal')
plt.legend()
plt.show()
```
