```python
import numpy as np
from scipy import stats
import pandas as pd

import sys
sys.path.insert(0, 'C:/Users/ander/Repositorios/val-metodos-analiticos/funcs')
import functions_paper as functions
```

```python
n_rep = 30
n_samples = 100000
n_samples = 5000

seeds = np.arange(1,n_samples+1)
rng = np.random.default_rng(42)
rng.shuffle(seeds)
```

## Normal distribution

```python
data_kurtosis = []
data_skew = []
plot_data = []
mean = []
std = []


for seed in seeds:
    normal_data = stats.norm.rvs(loc=0.0, scale=1.0, size=n_rep)
    plot_data.append(pd.DataFrame({"Data": normal_data, "seed": [seed]*len(normal_data)}))
    data_kurtosis.append(stats.kurtosis(normal_data)+3)
    data_skew.append(stats.skew(normal_data))
    mean.append(np.mean(normal_data))
    std.append(np.std(normal_data, ddof=1))
df = pd.concat(plot_data)
```

```python
functions.display_distribution_plots(df, data_skew, data_kurtosis, seeds, "norm")
```

```python
functions.display_skew_kurt_info(data_skew, data_kurtosis)
```
