# Dados gerados para exemplificar o probplot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

## Dados "Normais"

```python
n = 30
random_state = 42
```

```python
a = 0
x = stats.skewnorm.rvs(a, size=n, random_state=random_state)
df = pd.DataFrame({
    "dados": np.round(x, 5)
})
df.to_csv("probplot-normal.csv", index=False)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,5))
x_ordered = np.sort(x)
ax.plot(x_ordered, stats.skewnorm.pdf(x_ordered, a), alpha=0.6, label=f'a={a}')
plt.show()
```

```python
a = -8
x = stats.skewnorm.rvs(a, size=n, random_state=random_state)
df = pd.DataFrame({
    "dados": np.round(x, 5)
})
df.to_csv("probplot-left-skewed.csv", index=False)
```
