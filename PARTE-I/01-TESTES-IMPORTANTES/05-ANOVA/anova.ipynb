```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import random
from scipy.optimize import curve_fit
```

```python
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 11
```

```python
x1 = np.array([0.9023, 0.3697, 0.6821, 0.7556,])

x2 = np.array([0.4277, 0.7875, 0.5412, 0.5559,])

x3 = np.array([0.9030, 0.6939, 0.5660, 0.6280])

x4 = np.array([0.9147, 0.9720, 0.4580, 0.6176])

y = np.array([1, 1, 1, 1])
```

```python
fig, ax = plt.subplots(figsize=(7,5))
padding = .1
ax.scatter(x1, y, label="Grupo A")
ax.scatter(x2, y + padding, label="Grupo B")
ax.scatter(x3, y + padding + padding, label="Grupo C")
ax.scatter(x4, y + padding + padding + padding, label="Grupo D")

ax.set_xlim(-.2, 1.6)
ax.set_ylim(.8, 1.5)
ax.tick_params(axis='y', left=False, labelleft=False)
ax.legend(loc=0, bbox_to_anchor=(1.25, .75))
fig.tight_layout()
plt.show()
```
