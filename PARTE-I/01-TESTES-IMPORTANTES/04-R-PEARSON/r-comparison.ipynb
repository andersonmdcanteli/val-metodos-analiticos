## Cubica

### Gerando dados

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

cubic = np.poly1d(np.polyfit(x, y, 3))


x_data = []
y_data = []


for x in range(0,25,2):
    y = cubic(x)
    random.seed=42
    y = float(y + random.uniform(-5, 5))
    x_data.append(x)
    y_data.append(y)
```

```python




plt.figure(figsize=(6,4))
# dados experimentais
alfa = 0.8
color = 'k'
plt.scatter(x_data, y_data, facecolors='none', edgecolors=color, label=f'Data', alpha=alfa)


# regressão linear
reg = stats.linregress(x_data, y_data)
x_pred = np.array([min(x_data), max(x_data)])
y_pred = reg[0]*x_pred + reg[1]
label = None
label = "Linear"
plt.plot(x_pred, y_pred, c='r', label=label, alpha=alfa)


# cubico
cubico = np.polyfit(x_data, y_data, 3)
x_pred = np.linspace(min(x_data), max(x_data), 100)
y_pred = np.poly1d(cubico)
label = None
label = "Cúbico"
plt.plot(x_pred, y_pred(x_pred), c='b', label=label, alpha=alfa)


plt.ylim(50,125)
plt.legend(bbox_to_anchor=(1.25,1))
plt.show()
```

## Exponencial

```python
def func(x, a, b, c):
  #return a * np.exp(-b * x) + c
  return a * np.log(b * x) + c

x_data = np.linspace(1.1,5,50)   # changed boundary conditions to avoid division by 0
y_datay = func(x_data, 2.5, 1.3, 0.5)
y_data = y_datay + np.abs(0.2*np.random.normal(size=len(x_data)))

plt.figure(figsize=(6,4))
# dados experimentais
alfa = 0.8
color = 'k'
plt.scatter(x_data, y_data, facecolors='none', edgecolors=color, label=f'Data', alpha=alfa)




plt.legend(bbox_to_anchor=(1.25,1))
plt.show()
```
