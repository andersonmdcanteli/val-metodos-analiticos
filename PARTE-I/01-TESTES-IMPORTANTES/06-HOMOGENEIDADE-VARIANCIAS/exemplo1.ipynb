```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import sys
sys.path.insert(0, 'C:/Users/ander/Repositorios/val-metodos-analiticos/funcs')
import functions
import random
```

```python
x_setosa = np.array([5.1, 4.9, 4.7, 4.6, 5, 5.4, 4.6, 5, 4.4, 4.9])


x_versicolor = np.array([7, 6.4, 6.9, 5.5, 6.5, 5.7, 6.3, 4.9, 6.6, 5.2])

x_virginica = np.array([6.3, 5.8, 7.1, 6.3, 6.5, 7.6, 4.9, 7.3, 6.7, 7.2,])

df = pd.DataFrame({
    "setosa": x_setosa,
    "versicolor": x_versicolor,
    "virginica": x_virginica,
})

alpha = 0.05
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
df.boxplot(grid=False, ax=ax)
ax.set_ylabel("Comprimento das sépalas ($cm$)")

plt.show()
```

```python
levene = stats.levene(x_setosa, x_versicolor, x_virginica)
levene
```

```python
if levene.pvalue < alpha:
    print("Variâncias não homogêneas")
else:
    print("Variâncias homogêneas")
```

```python
functions.make_f_distribution_plot(levene.statistic, levene.pvalue, gl_numerator=df.shape[1]-1, gl_denominator=df.shape[0]*df.shape[1] - df.shape[1], alpha=alpha)
```
