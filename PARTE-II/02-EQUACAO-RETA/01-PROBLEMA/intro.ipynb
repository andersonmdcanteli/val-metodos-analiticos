```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 11
```

```python
x = np.array([0.1, 0.3, 0.5, 0.7, 0.9])

y = np.array([0.1, 0.15, 0.24, 0.3, 0.37])




x1 = [.1]*3
y1 = np.array([.1, .09, .112])

x2 = [.3]*3
y2 = np.array([.15, .14, .165])

x3 = [.5]*3
y3 = np.array([.24, .249, .228])

x4 = [.7]*3
y4 = np.array([.3, .29, .312])

x5 = [.9]*3
y5 = np.array([.37, .362, .385])


x_full = np.array(x1 + x2 + x3 + x4 + x5)
y_full = np.append(y1, [y2, y3, y4, y5])

reg = stats.linregress(x_full, y_full)

y_pred = reg[0]*x_full + reg[1]

fig, ax = plt.subplots(figsize=(6,5))
# ax.scatter(x, y, c=["b", "r", "orange", "k", "m"])

plt.scatter(x1, y1, c="b")
plt.scatter(x2, y2, c="r")
plt.scatter(x3, y3, c="orange")
plt.scatter(x4, y4, c="k")
plt.scatter(x5, y5, c="m")

plt.plot(x_full, y_pred, c='k', ls="--")


ax.spines['bottom'].set_color('k')
ax.spines['top'].set_color('k') 
ax.spines['right'].set_color('k')
ax.spines['left'].set_color('k')

ax.set_xlabel("Concentração do analíto")
ax.xaxis.label.set_color('k')
ax.set_ylabel("Sinal resposta")
ax.yaxis.label.set_color('k')
ax.tick_params(colors='k', axis="y", which='both')
ax.tick_params(colors='k', axis="x", which='both')

ax.set_xticks(x)
ax.set_xlim(0, 1)
ax.set_ylim(0, .45)
plt.show()
```
